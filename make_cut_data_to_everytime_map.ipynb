```python
import numpy as np
import pandas as pd
import os
import math
```

```python
df_column = []
for csv in os.listdir('./cutclean_raw_data'):
    df_column.append(csv[0:6])
print(df_column)
```

# construct everytime's map of open_price

```python
data_frame = []
for csv in os.listdir('./cutclean_raw_data'):
    data = pd.read_csv('./cutclean_raw_data/'+csv, encoding='utf-8', index_col='date')
    data_frame.append(data['open_price'])
print(len(data_frame))
print(type(data_frame))
```

```python
open_price = data_frame[0]
for i in range(1, len(data_frame)):
    open_price = pd.concat([open_price, data_frame[i]], axis=1, join='outer')
print(len(open_price))
```

```python
print(open_price.values)
```

```python
df_open_price = pd.DataFrame(open_price.values, index=open_price.index, columns=df_column)
print(df_open_price)
```

```python
df_open_price = df_open_price.fillna(method='pad')
print(df_open_price)
```

```python
df_open_price = df_open_price.fillna(method='bfill')
```

```python
print(df_open_price.shape)
```
