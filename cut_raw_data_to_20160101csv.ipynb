```python
for excel in os.listdir('./data_by_5min'):
    data = pd.read_excel('./data_by_5min/'+excel)
    excel_name = excel[0:6]+'.csv'
    data.columns = ['date', 'open_price', 'highest_price', 'lowest_price', 'close_price','turnover', 'volume']
    data = data.dropna(axis=0)
    data['date'] = pd.to_datetime(data['date'])
    data = data[data['date'] > pd.datetime(2016,1,1)]
    data.index = data['date']
    data = data.drop(['date'], axis=1)
    print(excel_name)
    data.to_csv('./cutclean_raw_data/'+excel_name)
```
